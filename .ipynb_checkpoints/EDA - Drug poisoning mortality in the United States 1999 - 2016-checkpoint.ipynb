```r
library(ggplot2)
library(dplyr)
```

## Drug poisoning mortality in the United States 1999 - 2016

```r
mortality <- read.csv('NCHS_-_Drug_Poisoning_Mortality_by_State__United_States.csv')
```

```r
colnames(mortality[colnames(mortality) == 'Race.and.Hispanic.Origin']) <- 'Race'
```

```r
mortality
```

```r
mortality %>%
filter(State == 'United States',
       Sex == 'Both Sexes',
       Age.Group == 'All Ages',
       Race.and.Hispanic.Origin == 'All Races-All Origins') %>%
ggplot(aes(x=Year, y = Crude.Death.Rate)) +
geom_point() +
ggtitle('Overdose death rates in the United States from 1999 to 2016')
```
