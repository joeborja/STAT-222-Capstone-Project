# EDA - Drug poisoning mortality by sex, age, race, and poverty

```r
library(ggplot2)
library(dplyr)
```

## Drug poisoning mortality in the United States 1999 - 2016

```r
mortality <- read.csv('../Data/NCHS_-_Drug_Poisoning_Mortality_by_State__United_States.csv')
```

Deaths per 100,000 people between 1999 and 2016 in the US.

```r
mortality %>%
filter(State == 'United States',
       Sex == 'Both Sexes',
       Age.Group == 'All Ages',
       Race.and.Hispanic.Origin == 'All Races-All Origins') %>%
ggplot(aes(x=Year, y = Crude.Death.Rate)) +
geom_point() +
ggtitle('Overdose death rates in the United States from 1999 to 2016')
```

Death rates between 1999 and 2016 by state:

```r
mortality %>%
filter(State != 'United States',
       Sex == 'Both Sexes',
       Age.Group == 'All Ages',
       Race.and.Hispanic.Origin == 'All Races-All Origins') %>%
ggplot(aes(x=Year, y = Crude.Death.Rate, color = State)) +
geom_line() +
ggtitle('Overdose death rates by state from 1999 to 2016') +
theme(legend.position = 'bottom')
```

National death rates by race

```r
mortality %>%
filter(State == 'United States',
       Sex == 'Both Sexes',
       Age.Group == 'All Ages') %>%
ggplot(aes(x=Year, y = Crude.Death.Rate, color = Race.and.Hispanic.Origin)) +
geom_line() +
ggtitle('Overdose death rates by race from 1999 to 2016') 
```

## Poverty data

The following data describe levels of poverty at the U.S. national level. These were taken from the U.S. Census Current Population Survey (CPS).

Poverty thresholds to roughly translate poverty to income levels may be found [here](https://www.census.gov/data/tables/time-series/demo/income-poverty/historical-poverty-thresholds.html).

```r
us_poverty <- read.csv('us_poverty.csv')
```

```r
head(us_poverty)
```

### All races

```r
mortality %>%
filter(State == 'United States',
       Sex == 'Both Sexes',
       Age.Group == 'All Ages',
       Race.and.Hispanic.Origin == 'All Races-All Origins') %>%
ggplot(aes(x=Year, y = Crude.Death.Rate)) +
geom_point() + 
ylim(0,30) +
geom_point(data = us_poverty %>% filter(Race == "ALL RACES"), aes(x = Year, y = below.poverty.percentage), color = "red") 
```

```r
aggregate_mortality <- mortality %>%
filter(State == 'United States',
       Sex == 'Both Sexes',
       Age.Group == 'All Ages',
       Race.and.Hispanic.Origin == 'All Races-All Origins')  

aggregate_poverty <- us_poverty %>%
filter(Race == "ALL RACES")
```

```r
aggregate_mortality
```

Correlation

```r
poverty_mortality <- merge(subset(aggregate_poverty[c("Year", "below.poverty.percentage")], Year != 2017),
aggregate_mortality[c("Year", "Crude.Death.Rate")], by="Year")

cor(poverty_mortality[c("below.poverty.percentage", "Crude.Death.Rate")])
```
